```python
import re
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load the log file
# file_path = Path("/mnt/HDD2/phu2/custom-guided-diffusion/train_with_batch/learning_curve.txt")
file_path = Path("/mnt/HDD2/phu2/custom-guided-diffusion/learning_curve_imagenet.txt")
with open(file_path, "r") as file:
    log_text = file.read()

# Extract epoch numbers and losses using regex
# matches = re.findall(r"epoch:\s+(\d+)\s+epoch_loss:\s+([0-9.]+)", log_text)
matches = re.findall(r"iteration\s+(\d+)\s+loss\s+([0-9.]+)", log_text)
epochs, losses = zip(*[(int(epoch), float(loss)) for epoch, loss in matches])


# Create a DataFrame for plotting
df = pd.DataFrame({"Epoch": epochs, "Loss": losses})

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df["Epoch"], df["Loss"], marker="o", linestyle="-")
plt.title("Epoch Loss over Time")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.grid(True)
plt.tight_layout()
plt.show()
```
