```python
from PIL import Image, ImageChops
import numpy as np
import matplotlib.pyplot as plt

def disparity_normalization( disp): # disp is an array in uint8 data type
        # disp_norm = cv2.normalize(src=disp, dst= disp, beta=0, alpha=255, norm_type=cv2.NORM_MINMAX)
        _min = np.amin(disp)
        _max = np.amax(disp)
        #disp_norm = disp - _min * 255.0 / (_max - _min)
        disp_norm = (disp - _min) * 255.0 / (_max - _min)
        disp_norm = np.uint8(disp_norm)

        return disp_norm 


with_refine = Image.open('/mnt/HDD2/phudh/custom-guided-diffusion/visualize/[with refine noise] random_sample_x_0_t_minus_1_at_190_sample_from_x_0.png')
without_refine = Image.open('/mnt/HDD2/phudh/custom-guided-diffusion/visualize/[without refine noise] random_sample_x_0_t_minus_1_at_190_sample_from_x_0.png')

difference = ImageChops.difference(with_refine, without_refine)
difference = np.array(difference)
print(f'difference: min {difference.min()} max {difference.max()} mean {difference.mean()}')
# difference = disparity_normalization(difference)
plt.imshow(difference)
```
